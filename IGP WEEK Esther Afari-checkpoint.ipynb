```python
import pandas as pd
```

```python
product_sales = [{'product_id':23, 'name':'computer', 'wholesale_price': 500, 'retail_price':1000, 'sales':100},                                           
                 {'product_id':96, 'name':'Python Workout', 'wholesale_price': 35,'retail_price':75, 'sales':1000},
                 {'product_id':97, 'name':'Pandas Workout', 'wholesale_price': 35, 'retail_price':75, 'sales':500},
                 {'product_id':15, 'name':'banana', 'wholesale_price': 0.5,'retail_price':1, 'sales':200},
                 {'product_id':87, 'name':'sandwich', 'wholesale_price': 3,'retail_price':5, 'sales':300}]
```

```python
product_sales_df = pd.DataFrame(product_sales)
product_sales_df 
```

```python
product_sales_df['net_revenue_per_product'] = (product_sales_df ['retail_price'] - product_sales_df ['wholesale_price']) * product_sales_df ['sales']
product_sales_df
```

```python
total_net_revenue = product_sales_df['net_revenue_per_product'].sum()
total_net_revenue
```

```python
product_sales_df[(product_sales_df ['wholesale_price']*2) < product_sales_df['retail_price']]
```

```python
product_dict = {
                'computer' : 'computer', 
                'banana' : 'Food',
                'sandwich' : 'Food',
                'Pandas Workout' : 'Books',
                'Python Workout' : 'Books'
}
print(product_dict)


#Assign product name to category
product_sales_df['Category'] = product_sales_df['name'].map(product_dict)

#Total revenue for each category
revenue_for_the_category = product_sales_df.groupby('Category')['net_revenue_per_product'].sum().reset_index()
print(revenue_for_the_category)
```

```python
# Applying 30% discount to the wholesale price
product_sales_df['Discounted_Wholesale_Price'] = product_sales_df['wholesale_price'] * 0.7

# Net revenue for each product with the discounted wholesale price
product_sales_df['Net Revenue with Discount'] = (product_sales_df['retail_price'] - product_sales_df['Discounted_Wholesale_Price']) * product_sales_df['sales']
product_sales_df
```

```python
# Calculate the current income and income after applying each tax rate
product_sales_df['Tax 15%']= 0.15 * product_sales_df['Net Revenue with Discount']
product_sales_df[' Tax 20%'] =  0.20 * product_sales_df['Net Revenue with Discount']
product_sales_df['Tax 25%']  = 0.25 * product_sales_df['Net Revenue with Discount']
    
# Display the DataFrame with the added columns
product_sales_df
```
